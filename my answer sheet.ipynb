#Question 1: A company stores monthly revenue data that should not be modified accidentally.
1. Would you use a list or tuple? Justify.
2. Create a tuple of quarterly revenues.
3. Try modifying one value. What happens? Explain why.

#Would you use a list or tuple? Justify:

I would use a tuple. The requirement explicitly states that the monthly revenue data should not be modified accidentally. Tuples in Python are immutable, meaning once they are created, their elements cannot be changed, added, or removed. Lists are mutable and prone to accidental modification.


#& 3. Modifying a tuple:

 If you attempt to assign a new value to a specific index in a tuple, Python will raise a TypeError because the object does not support item assignment.  

```python
# 2. Create a tuple of quarterly revenues
quarterly_revenues = (150000, 185000, 160000, 210000)
print(f"Quarterly Revenues: {quarterly_revenues}")

# 3. Try modifying one value (This will intentionally throw an error)
try:
    quarterly_revenues = 160000
except TypeError as e:
    print(f"Error caught: {e}")
    print("Explanation: Tuples are immutable, so their values cannot be reassigned after creation.")
```

#Question 2: You are given student performance data:


students = {


"Amit": 85,


"Sara": 92,


"John": 78,


"Meera": 88


}


Tasks:
1. Add a new student.
2. Update marks of one student.
3. Find the student with the highest marks.
4. Convert dictionary keys into a list.
5. Explain why dictionaries are important in Data Science workflows.

```python
# Given student performance data
students = {
    "Amit": 85,
    "Sara": 92,
    "John": 78,
    "Meera": 88
} # [cite: 242, 243, 244, 245, 246, 247]

# 1. Add a new student
students["Rahul"] = 95 # [cite: 249]

# 2. Update marks of one student
students["Amit"] = 89 # [cite: 252]

# 3. Find the student with the highest marks
# max() looks at the keys by default, so we pass key=students.get to evaluate based on values
top_student = max(students, key=students.get) # [cite: 253]
print(f"Student with highest marks: {top_student} ({students[top_student]})")

# 4. Convert dictionary keys into a list
student_names_list = list(students.keys()) # [cite: 254]
print(f"List of student names: {student_names_list}")
```

#(5)Why dictionaries are important in Data Science workflows:



 Dictionaries store data in key-value pairs. This structure perfectly mirrors how real-world data is often transmitted (like JSON files from web APIs). In workflows, they are highly efficient for looking up data without needing to know its index position ($O(1)$ time complexity), and they are frequently used to map categorical variables to numerical features before training machine learning models

#Question 3:  An e-commerce platform has customer IDs from two campaigns:


campaign1 = {101, 102, 103, 104, 105}


campaign2 = {104, 105, 106, 107}
1. Find unique customers across both campaigns.
2. Find customers common in both.
3. Find customers exclusive to campaign1.
4. Explain how sets help in removing duplicates in real DS projects.

```python
# Given campaign data
campaign1 = {101, 102, 103, 104, 105} # [cite: 258]
campaign2 = {104, 105, 106, 107} # [cite: 259]

# 1. Find unique customers across both campaigns (Union)
unique_all = campaign1.union(campaign2) # [cite: 260]
print(f"Unique customers across both: {unique_all}")

# 2. Find customers common in both (Intersection)
common_customers = campaign1.intersection(campaign2) # [cite: 261]
print(f"Common customers: {common_customers}")

# 3. Find customers exclusive to campaign1 (Difference)
exclusive_to_c1 = campaign1.difference(campaign2) # [cite: 262]
print(f"Exclusive to Campaign 1: {exclusive_to_c1}")
```

#How sets help in removing duplicates in real DS projects:

In real-world data, you often end up with duplicate rows or values (e.g., a user clicking an ad multiple times). Because a Python set is a mathematical collection of unique, non-repeating elements, simply converting a list to a set automatically filters out all duplicates instantly. It is the fastest way to find unique categories or overlapping cohorts.

#Question 4: You receive raw CSV data where:


price = "499.99"


quantity = "5"
1. Convert them to correct numeric types.
2. Compute total revenue.
3. Explain why typecasting errors can break ML pipelines.

```python
# Given raw CSV data
price_raw = "499.99" # [cite: 268]
quantity_raw = "5" # [cite: 268]

# 1. Convert them to correct numeric types
price = float(price_raw) # [cite: 269]
quantity = int(quantity_raw) # [cite: 269]

# 2. Compute total revenue
total_revenue = price * quantity # [cite: 270]

print(f"Converted Price: {price} (Type: {type(price).__name__})")
print(f"Converted Quantity: {quantity} (Type: {type(quantity).__name__})")
print(f"Total Revenue: {total_revenue}")
```

#Why typecasting errors can break ML pipelines:

Machine Learning algorithms (like those in Scikit-Learn or neural networks in PyTorch) operate entirely on mathematical equations. They cannot multiply or optimize a string like "499.99". If raw CSV text is not explicitly typecasted into floats or integers (numerical formats), the mathematical operations in the pipeline will crash with a TypeError.

#Question 5:   Given:


sales = [12000, 15000, 11000, 18000, 17000]


Using a loop:
1. Calculate total revenue.
2. Count how many months had sales above 15000.
3. Find maximum sales without using built-in max().

```python
sales = [12000, 15000, 11000, 18000, 17000] # [cite: 274]

# Initialize variables for the loop constraints
total_revenue = 0
count_above_15000 = 0
max_sales = sales[0] # Assume first month is highest initially

# Using a single loop for all 3 tasks [cite: 275]
for month_sale in sales:
    # 1. Calculate total revenue [cite: 276]
    total_revenue += month_sale

    # 2. Count how many months had sales above 15000 [cite: 277]
    if month_sale > 15000:
        count_above_15000 += 1

    # 3. Find maximum sales without using built-in max() [cite: 278]
    if month_sale > max_sales:
        max_sales = month_sale

print(f"Total Revenue: {total_revenue}")
print(f"Months with sales > 15000: {count_above_15000}")
print(f"Maximum Sales: {max_sales}")
```

Question 6: You are analyzing product ratings:


ratings = [


[5, 4, 3],


[4, 4, 5],


[3, 5, 4]


]


1. Compute average rating.
2. Print all ratings greater than 4.
3. Explain how nested loops are used in matrix-based ML computations.

```python
# Given product ratings
ratings = [
    [5, 4, 3],
    [4, 4, 5],
    [3, 5, 4]
] # [cite: 284, 285, 286, 287, 288]

total_sum = 0
total_count = 0

print("Ratings greater than 4:")

# Iterating through the nested list
for product in ratings:
    for rating in product:
        # 1. Accumulate values to compute average rating [cite: 289]
        total_sum += rating
        total_count += 1

        # 2. Print all ratings greater than 4 [cite: 290]
        if rating > 4:
            print(rating)

average_rating = total_sum / total_count
print(f"\nAverage Rating: {average_rating:.2f}")
```

#How nested loops are used in matrix-based ML computations:

 Matrices (like images, which are grids of pixels, or structured datasets of rows and columns) are intrinsically multi-dimensional. To process them, ML algorithms must traverse the data structure using nested loops: an outer loop to iterate through the rows, and an inner loop to iterate through the individual columns/features within that row. This is the foundation of matrix multiplication and operations like convolutional layers in computer vision.

This makes the code readable, easily testable, and allows you to fix a bug in one place rather than hunting it down across thousands of lines of code

#Question 8: Given


prices = [500, 1500, 2500, 800, 1200]


1. Use map() + lambda to add 18% GST to all prices.
2. Use filter() to select prices above 1000.
3. Explain where such transformations are used in real-world pipeline

```python
# Given prices
prices = [500, 1500, 2500, 800, 1200] # [cite: 303]

# 1. Use map() + lambda to add 18% GST (price * 1.18) [cite: 304]
# We use round() to keep it clean
prices_with_gst = list(map(lambda p: round(p * 1.18, 2), prices))
print(f"Prices with 18% GST: {prices_with_gst}")

# 2. Use filter() to select prices above 1000 [cite: 305]
# Using the original prices list
high_prices = list(filter(lambda p: p > 1000, prices))
print(f"Original prices above 1000: {high_prices}")
```

#Where such transformations are used in real-world pipelines:

 These functional programming concepts are the backbone of data transformation tasks. map() is conceptually identical to Pandas' .apply() method, used to apply a formula (like normalizing data or scaling features) across an entire dataset column simultaneously. filter() is used extensively in data cleaning to drop invalid rows, remove missing values, or isolate specific cohorts before feeding the data into a machine learning model.

#Question 9: Write a generator that yields squares of numbers from 1 to 1,000,000.


Explain difference between:


● List comprehension


● Generator expression


Why are generators preferred in big data scenarios?

#List Comprehension vs. Generator Expression:

 A list comprehension computes every single value immediately and stores the entire resulting list in your computer's RAM.

  A generator expression, however, evaluates "lazily." It doesn't compute the values upfront; instead, it generates one value at a time only when specifically requested.

#Why generators are preferred in big data scenarios:

If you have a dataset with billions of rows, a list comprehension will try to load all billions of rows into memory at once, crashing your system (Out Of Memory error). Generators solve this by streaming the data one piece at a time, keeping memory usage exceptionally low regardless of how massive the dataset is.

```python
# Write a generator that yields squares of numbers from 1 to 1,000,000 [cite: 312]
def square_generator():
    for i in range(1, 1000001):
        yield i ** 2

# Alternatively, writing it as a generator expression:
# squares_gen = (i ** 2 for i in range(1, 1000001))

# Let's test it by getting just the first 5 values using next()
gen = square_generator()
print("First 5 squares generated efficiently:")
for _ in range(5):
    print(next(gen))
```

#Question 10: Design a class:


class Product:


With:


● Attributes: name, price, quantity


● Method: calculate_revenue()
Create 3 objects and print their revenue.


```python
# Design a class [cite: 318, 319, 320, 321, 322]
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def calculate_revenue(self):
        return self.price * self.quantity

# Create 3 objects [cite: 323]
product1 = Product("Smartwatch", 250, 10)
product2 = Product("Headphones", 150, 30)
product3 = Product("Powerbank", 40, 100)

# Print their revenue [cite: 323]
print(f"Revenue for {product1.name}: ${product1.calculate_revenue()}")
print(f"Revenue for {product2.name}: ${product2.calculate_revenue()}")
print(f"Revenue for {product3.name}: ${product3.calculate_revenue()}")
```
